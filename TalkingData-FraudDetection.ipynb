```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_sample = pd.read_csv('Data/train_sample.csv', parse_dates=True)
train_sample.head()
```

```python
clicktime = pd.DataFrame(train_sample, columns = ['click_time'])
clicktime.head()
```

```python
clicktime.dtypes
```

```python
clicktime['click_time'] = clicktime['click_time'].astype('datetime64[ns]')
```

```python
unique_clicks = pd.Series(clicktime['click_time'].value_counts())
```

```python
unique_clicks
```

```python
unique_clicks.plot(kind='line', figsize=(20,10))
plt.show()
```

```python
subset_highest_day = unique_clicks['2017-11-08 00:00:00':'2017-11-08 23:59:59']
subset_highest_day.head()
```

```python
subset_highest_day.plot(kind='line', figsize=(20,10))
plt.show()
```

```python
subset_highest_hour = unique_clicks['2017-11-08 12:00:00':'2017-11-08 13:00:00']
subset_highest_hour.head()
```

```python
subset_highest_hour.plot(kind='line', figsize=(20,10))
plt.show()
```

```python
subset_highest_10mins = unique_clicks['2017-11-08 12:00:00':'2017-11-08 12:10:00']
subset_highest_10mins.head()
```

```python
subset_highest_10mins.plot(kind='line', figsize=(20,10))
plt.show()
```
